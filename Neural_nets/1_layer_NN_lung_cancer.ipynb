```python
def sigmoid(Z) :
    Z = np.asarray(Z, dtype = np.float64)
    return(1/(1+np.exp(-Z)))
```

```python
def initialize_parameters(n_x, n_h, n_y):
    """
    Argument:
    n_x -- size of the input layer
    n_h -- size of the hidden layer
    n_y -- size of the output layer
    
    Returns:
    params -- python dictionary containing your parameters:
                    W1 -- weight matrix of shape (n_x,n_h)
                    b1 -- bias vector of shape (1,n_h)
                    W2 -- weight matrix of shape (n_h,n_y)
                    b2 -- bias vector of shape (1,n_y)
    """
    
    W1 = np.random.randn(n_x,n_h) * 0.01
    # (15,5)
    b1 = np.zeros((1,n_h))
    # (1,5)

    W2 = np.random.randn(n_h,n_y) * 0.01
    # (5,1)
    b2 = np.zeros((1,n_y))
    # (1,1)
    
    parameters = {"W1": W1,
                  "b1": b1,
                  "W2": W2,
                  "b2": b2}
    
    return parameters
```

```python
def cost_function(y,A2):
    cost =  -np.mean(y*np.log(A2) + (1-y)*np.log(1-A2))
    return cost
```

```python
def forward_propagation(X, parameters) :

    W1 = parameters["W1"]
    b1 = parameters["b1"]
    W2 = parameters["W2"]
    b2 = parameters["b2"]

    Z1 = np.dot(X,W1) + b1
    # 220,5
    A1 = sigmoid(Z1).reshape(X.shape[0],b1.shape[1])
    
    Z2 = np.dot(A1,W2) + b2
    # 220,1
    A2 = sigmoid(Z2).reshape(X.shape[0],b2.shape[1])
    
    cache = {"Z1": Z1,
             "A1": A1,
             "Z2": Z2,
             "A2": A2}
    return cache
```

```python
def backward_propagation(cache,parameters,X,y,m):
    
    W1 = parameters["W1"]
    b1 = parameters["b1"]
    W2 = parameters["W2"]
    b2 = parameters["b2"]

    A1 = cache["A1"]
    A2 = cache["A2"]
    Z1 = cache["Z1"]
    Z2 = cache["Z2"]

    dZ2 = A2 - y
    dW2 = (1/m)*(np.dot(A1.T,dZ2))
    db2 = np.mean(dZ2)

    dZ1 = np.dot(dZ2,W2.T)*np.multiply(A1,(A1-1))
    dW1 = (np.dot(X.T,dZ1))
    db1 = (1/m) *(np.sum(dZ1, axis=0, keepdims=True))

    grads = {
        "dW1" : dW1,
        "db1" : db1,
        "dW2" : dW2,
        "db2" : db2
    }

    return grads
```

```python
def update_parameters(parameters, grads, learning_rate):

    W1 = parameters["W1"]
    b1 = parameters["b1"]
    W2 = parameters["W2"]
    b2 = parameters["b2"]

    dW1 = grads["dW1"]
    db1 = grads["db1"]
    dW2 = grads["dW2"]
    db2 = grads["db2"]

    W1 = W1 - learning_rate * dW1
    b1 = b1 - learning_rate * db1
    W2 = W2 - learning_rate * dW2
    b2 = b2 - learning_rate * db2
    
    parameters = {"W1": W1,
                  "b1": b1,
                  "W2": W2,
                  "b2": b2}
    
    return parameters
```

```python
   
n_x = 15
n_y = 1
n_h = 10
iterations = 10000
learning_rate = 0.01
m = X_train.shape[0]

parameters = initialize_parameters(n_x,n_h,n_y)
for i in range(0, iterations) :
    cache = forward_propagation(X_train,parameters)

    grads = backward_propagation(cache,parameters,X_train,y_train,m)

    parameters = update_parameters(parameters, grads, learning_rate)
    if i%500 == 0:
        print("Grads shapes:\n dW1, db1, dW2, db2",grads["dW1"].shape, grads["db1"].shape, grads["dW2"].shape, grads["db2"].shape)
        print("Cost for",i,"iterations",cost_function(y_train, cache["A2"]))
        print(cache["A2"][:5])
```

```python
result = forward_propagation(X_test,parameters)
final = result["A2"]
final[final > 0.5] = 1
final[final < 0.5] = 0

print(np.mean(final == y_test))
print(final,y_test)
```
